### Data cleaning

```python
df.isnull().sum()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df.fillna(method='pad', inplace=True)
```

```python
df.isnull().sum()
```

### Separate input and output data

```python
x = df.drop('CLASS', axis = 1)
y = df['CLASS']
```

```python
sns.countplot(x = df['SHAPE']);
```

```python
sns.countplot(x = df['SOIL_TYPE']);
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
x['SHAPE'] = le.fit_transform(x['SHAPE'])
x['SOIL_TYPE'] = le.fit_transform(x['SOIL_TYPE'])
```

```python
x.dtypes
```

### Outlier Detection and Removal

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(contamination= 0.05, random_state= 0)
```

```python
clean = iso.fit_predict(x, y)
```

```python
x_unclean = x[clean == -1]
```

```python
x_unclean.shape
```

```python
x_clean = x[clean == 1]
y_clean = y[clean == 1]
```

```python
x_clean.shape, y_clean.shape
```

### Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
skf = SelectKBest(k = 5, score_func = chi2)
```

```python
skf.fit(x_clean, y_clean)
```

```python
skf.get_support()
```

```python
x_clean = x_clean.iloc[:,skf.get_support()]
```

```python
x_clean
```

### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x_clean)
```

### Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_clean,
                                                    random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

### Build the model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(random_state= 0)
```

```python
clf.fit(x_train, y_train)
```

### Performance Analysis

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
new = [[120.78, 3.25, 1123, 22.34, 30.11]]
```

```python
clf.predict(scaler.transform(new))
```
