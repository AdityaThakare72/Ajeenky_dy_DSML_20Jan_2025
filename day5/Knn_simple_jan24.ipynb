```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('ingredients.csv')
df.head()
```

```python
sns.jointplot(x = 'crunchiness', y = 'sweetness', data = df, hue = 'class');
```

```python
x = df[['crunchiness','sweetness']]
y = df['class']
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 4)
```

```python
knn.fit(x, y)
```

```python
tomato = [[4, 6]]
```

```python
knn.predict(tomato)
```
